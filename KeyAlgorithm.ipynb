```python
%matplotlib notebook

import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import rbf_kernel
import numpy as np
m = 100
x = np.linspace(-5, 5, m)
xr = x.reshape(-1, 1)
y = np.sinc(x)
plt.plot(xr, y, '.r')
```

```python
%matplotlib notebook
K = rbf_kernel(xr, xr, 0.001)
I = np.identity(m)
gamma = 100
const = m*I + K
solution = np.linalg.solve(const, y).reshape(-1, 1)
Ksum = np.sum(K, axis=1).reshape(-1,1)
yy = np.multiply(solution, Ksum)
plt.plot(xr, y, '.r')
plt.plot(xr, yy, '.b')
```
